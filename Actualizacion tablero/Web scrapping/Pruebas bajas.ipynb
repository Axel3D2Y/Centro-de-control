```python
import pandas as pd 
from datetime import datetime

```

```python
df = pd.read_excel(r"C:\Users\centro.control\Downloads\CATALOGO AGOSTO SEM 36 OK.xlsx")
```

```python
df["FECHA ALTA"] = pd.to_datetime(df["FECHA ALTA"], errors="coerce")
df["FECHA BAJA"] = pd.to_datetime(df["FECHA BAJA"], errors="coerce")
```

```python
df["Dias laborados"] = (df["FECHA BAJA"] - df["FECHA ALTA"]).dt.days
```

```python
df["FECHA ALTA"] = df["FECHA ALTA"].dt.date
df["FECHA BAJA"] = df["FECHA BAJA"].dt.date
```

```python
df_bajas = df[df['ESTATUS']== 'B']
```

```python
df_bajas.groupby(by='SEXO')['Dias laborados'].mean()
```
