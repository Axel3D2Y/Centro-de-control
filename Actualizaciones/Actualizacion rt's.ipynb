```python
import pandas as pd 
import numpy as np
```

```python
rts_2= pd.read_excel(r"C:\Users\centro.control\Downloads\Rts-Sem-44-2025.xlsx")
sitios_sap = pd.read_excel(r"C:\Users\centro.control\Desktop\Monitor CMC\Sitios SAP.xlsx")
```

```python
#Actualizar cada semana con las fechas correspondientes  de Domingo a Sábado
diccionariofechas= {
    'Domingo': 45942,
    'Lunes': 45943,
    'Martes': 45944,
    'Miercoles': 45945,
    'Jueves': 45946,
    'Viernes': 45947,
    'Sabado': 45948,
}
```

```python
def rts_dia(dia):
    rts_x_dia = rts_2[rts_2[dia].notna() ][ ['Nombre', 'Sitio donde acude', 'Turno'] ]
    rts_x_dia['Dia semana'] = dia
    return rts_x_dia

def dia_a_fecha(dia):
    return diccionariofechas.get(dia, 'fecha no encontrada')

dias_semana = ['Domingo', 'Lunes', 'Martes', 'Miercole', 'Jueves', 'Viernes', 'Sabado']
```

```python
rts_limpios = pd.DataFrame()
for dia in dias_semana:
    aux = []
    aux = pd.DataFrame(rts_dia(dia))
    rts_limpios = pd.concat([rts_limpios, aux], ignore_index=True)
```

```python
rts_limpios['fecha'] = rts_limpios['Dia semana'].apply(dia_a_fecha)
rts_limpios['Nomenclatura'] = 'RT'
rts_limpios['Asistencia']= 1
cols = [
    'candidate_id', 'NumWeek', 'employment',
    'incidencia', 'Significado','Tipo'
]
rts_limpios[cols] = np.nan
```

```python
sitios_sap
```

```python
rts_limpios= pd.merge(
    rts_limpios,
    sitios_sap[['NOMBRE COMERCIAL', 'SITIO', 'id']],
    left_on='Sitio donde acude',
    right_on='NOMBRE COMERCIAL',
    how='left'
)
```

```python
rts_limpios
```

```python
rts_limpios = rts_limpios[['id',
                    'candidate_id',
                    'SITIO',
                    'Nombre',
                    'employment',
                    'Turno',
                    'NumWeek',
                    'incidencia',
                    'fecha',
                    'Nomenclatura',
                    'Significado',
                    'Asistencia',
                    'Tipo']]
```

```python
rts_limpios.rename(columns={
    '': 'sitio',
    'Nombre': 'nombre_completo',
    'Turno': 'turno'})
```
