```python
import os
import pandas as pd
import glob
```

```python
folder_path ="C:\\Users\\centro.control\\Documents\\Python\\Proyectos\\Rutinas actuales vs frecuencia\\Archivos xlxs"
file_pattern = os.path.join(folder_path, '*.xlsx')
```

```python
dfs = []
for file in glob.glob(file_pattern):
    df = pd.read_excel(file)
    dfs.append(df)
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
combined_df .columns = combined_df .iloc[0]
combined_df  = combined_df .drop(index=[0])
combined_df .reset_index(drop=True, inplace=True)
```
