```python
import pandas as pd 
import numpy as np
```

```python
data= pd.read_excel(r"C:\Users\centro.control\Documents\Python\Vacantes.xlsx")
```

```python
data['Fecha del reporte'] = data['Fecha del reporte'].dt.strftime("%Y-%m-%d")
```

```python
data_vacantes = data[['Sitio','Turno','Fecha del reporte','Vacante']]
data_vacantes
```

```python
tabla_pivot = data_vacantes.pivot(
    index=['Sitio', 'Turno'],
    columns='Fecha del reporte',
    values='Vacante'
)
tabla_pivot= tabla_pivot.fillna(0)
tabla_pivot= tabla_pivot.reset_index()
fechas = tabla_pivot.columns[2:]
```

```python
vacantes_nuevas={}
vacantes_nuevas['SITIOS'] = tabla_pivot['Sitio']
vacantes_nuevas['Turno'] = tabla_pivot['Turno']
for i in range(len(fechas)-1):
    vacantes_nuevas[f'Nuevas: {fechas[i+1]} - {fechas[i]}'] = (tabla_pivot[fechas[i+1]] - tabla_pivot[fechas[i]]).clip(lower=0)
```

```python
vacantes_cubiertas={}

vacantes_cubiertas['SITIOS'] = tabla_pivot['Sitio']
vacantes_cubiertas['Turno'] = tabla_pivot['Turno']
for i in range(len(fechas)-1):
    vacantes_cubiertas[f'Cubiertas: {fechas[i+1]} - {fechas[i]}'] = (tabla_pivot[fechas[i]] - tabla_pivot[fechas[i+1]]).clip(lower=0)
```

```python
vacantes_sin_movimiento={}
for i in range(len(fechas)-1):
    vacantes_sin_movimiento[f'Nuevas: {fechas[i+1]} - {fechas[i]}'] = tabla_pivot[[fechas[i], fechas[i+1]]].min(axis=1)
```

```python
vacantes_nuevas=pd.DataFrame(vacantes_nuevas)
vacantes_cubiertas=pd.DataFrame(vacantes_cubiertas)
vacantes_sin_movimiento=pd.DataFrame(vacantes_sin_movimiento)
```

```python
print(np.sum(vacantes_cubiertas.iloc[:,-1]).astype(int))
print(np.sum(vacantes_nuevas.iloc[:,-1]).astype(int))
print(np.sum(vacantes_sin_movimiento.iloc[:,-1]).astype(int))
```

```python
def dias_vacantes(fila):
    contador = 0
    for val in  reversed(fila):
        if val != 0:
            contador += 1
        else:
            break
    return contador  
```

```python
conteos = vacantes_sin_movimiento.apply(dias_vacantes, axis=1)
```

```python
conteos = [x+1 if x != 0 else 0 for x in conteos]
```

```python

vacantes_sin_movimiento['SITIOS'] = tabla_pivot['Sitio']
vacantes_sin_movimiento['Turno'] = tabla_pivot['Turno']
vacantes_sin_movimiento['dias con vacantes'] = conteos

vacantes_sin_movimiento = vacantes_sin_movimiento.sort_values(by=['dias con vacantes'], ascending=False)
```
